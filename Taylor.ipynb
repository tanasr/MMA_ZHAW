```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
q = 0.5
s = 0
print(f"s                    k\n")
for k in range(0,30):
    s = s+q**k
    print(f'{s:<.9f} {k:>10}')
print(f'\nvia formula: {1/(1-q)}')
```

Darstellung, wie die Koeffizienten c0, c1 und c2 das Polynom platzieren, ausrichten und formen

```python
x = np.linspace(-2*np.pi,2*np.pi,1000)
f = lambda x: c0 + c1*x + c2*x**2
plt.plot(x,np.cos(x),'k')
c0,c1,c2 = 0.5,0.2,-0.2
plt.plot(x, f(x),label=f'$c_0$: {c0}, $c_1$: {c1}, $c_2$: {c2}',lw=1)
c0,c1,c2 = 0.4,0.0,-0.05
plt.plot(x, f(x),label=f'$c_0$: {c0}, $c_1$: {c1}, $c_2$: {c2}',lw=1)
c0,c1,c2 = 1.0,0.02,0.2
plt.plot(x, f(x),label=f'$c_0$: {c0}, $c_1$: {c1}, $c_2$: {c2}',lw=1)
plt.legend(fontsize='8')
plt.grid()
ax = plt.gca()
ax.set_ylim([-2,2])
plt.show()
```

Darstellung wie der Koeffizient c2 das Polynom formt (gegeben sind c0 und c1, die bleiben konstant)

```python
c0 = 1; c1 = 0
x = np.linspace(-2*np.pi,2*np.pi,1000)
P2 = lambda x: c0 + c1*x + c2*x**2
for c2 in np.arange(-1,1,0.45):
    plt.plot(x,np.cos(x),'k')
    plt.plot(x, P2(x),label=f'{c0},{c1},{c2}',lw=1)
ax = plt.gca()
ax.set_ylim([-2,2])
plt.grid()
plt.show()
```

Nun bilden wir das Taylor Polynom 2. Grades um x = 0 der funktion cos(x)


```python
c0 = 1; c1 = 0; c2 = -1/2
x = np.linspace(-np.pi,np.pi,1000)
P2 = lambda x: c0 + c1*x + c2*x**2

plt.plot(x,np.cos(x),'k')
plt.plot(x, f(x),label=f'{c0},{c1},{c2}',lw=1)
ax = plt.gca()
ax.set_ylim([-1.25,1.25])
plt.legend()
plt.grid()
plt.show()


# berprfe mathematisch den wert der echten funktion bei x = 0.22 und des polynoms bei x = 0.22
x0 = 0.22
true = np.cos(x0)
approx = P2(x0)
err = np.abs(true-approx)
print(f'true value: {true:.5f}, approx value: {approx:.5f}, error: {err:.5f}')
```

Die Approximation ist sehr gut wenn wir uns nicht weit entfernen vom zentralen punkt x = 0. Bei der Stelle x = 1 erkennt man bereits eine Abweichung, und je weiter wir uns wegbewegen von x = 0, desto schlechter entspricht das Taylor Polynom der Zielfunktion. Bei x = 2 ist die Abweichung zur echten Funktion sehr gross, das heisst das Polynom kann nicht mehr als gute Approximation genommen werden

```python
x0 = 1
true = np.cos(x0)
approx = P2(x0)
err = np.abs(true-approx)
print(f'true value: {true:.5f}, approx value: {approx:.5f}, error: {err:.5f}')


x0 = 2
true = np.cos(x0)
approx = P2(x0)
err = np.abs(true-approx)
print(f'true value: {true:.5f}, approx value: {approx:.5f}, error: {err:.5f}')
```

Natrlich kann das Polynom vebessert werden indem man Terme von hherer Ordnung hinzunehmen. Bspw. kann eine bessere Approximation durch das Hinzufgen von den weiteren Koeffizienten (c3) und c4 vebessert werden:


```python
c0 = 1; c1 = 0; c2 = -1/2; c3 = 0; c4 = 1/24
x = np.linspace(-np.pi,np.pi,1000)
P4 = lambda x: c0 + c1*x + c2*x**2 + c3*x**3 + c4*x**4

plt.plot(x,np.cos(x),'k')
plt.plot(x,P2(x),lw=1, label=r'$P_2(x)$')
plt.plot(x, P4(x),lw=1, label=r'$P_4(x)$')
ax = plt.gca()
ax.set_ylim([-1.25,1.25])
plt.grid()
plt.show()
```

Vergleichen wir nun die beiden Polynome, sieht man deutlich wie das Polynom P4 (4. Grades) die Funktion cos(x) um den Ursprung x = 0 besser approximiert. So ist auch gut ersichtlich, wie das Polynom bei x = 1 nher an cos(x) liegt und somit an dieser Stelle einen genaueren Wert liefert: 

```python

x0 = 1
true = np.cos(x0)
approxP2 = P2(x0)
approxP4 = P4(x0)
errP2 = np.abs(true-approxP2)
errP4 = np.abs(true-approxP4)
print(f'true value: {true:.5f}, P2 approx: {approxP2:.5f}, error: {errP2:.5f}')
print(f'true value: {true:.5f}, P4 approx: {approxP4:.5f}, error: {errP4:.5f}')
```

```python
# noch deutlicher wird es weiter vom Ursprung, bei x = 2 ist das Polynom wesentlich genauer
x0 = 2
true = np.cos(x0)
approxP2 = P2(x0)
approxP4 = P4(x0)
errP2 = np.abs(true-approxP2)
errP4 = np.abs(true-approxP4)
print(f'true value: {true:.5f}, P2 approx: {approxP2:.5f}, error: {errP2:.5f}')
print(f'true value: {true:.5f}, P4 approx: {approxP4:.5f}, error: {errP4:.5f}')
```

Um eine noch bessere Annherung zu erhalten, muss das Polynom erweitert werden durch weitere Koeffizienten


```python
c0 = 1; c1 = 0; c2 = -1/2; c3 = 0; c4 = 1/24; c5 = 0; c6 = -1/720
x = np.linspace(-np.pi,np.pi,1000)
P6 = lambda x: c0 + c1*x + c2*x**2 + c3*x**3 + c4*x**4 + c5*x**5 + c6*x**6

plt.plot(x,np.cos(x),'k', label='exact')
plt.plot(x,P2(x),lw=1, label=r'$P_2(x)$')
plt.plot(x,P4(x),lw=1, label=r'$P_4(x)$')
plt.plot(x,P6(x),lw=1, label=r'$P_6(x)$')
ax = plt.gca()
ax.set_ylim([-1.5,1.5])
plt.legend()
plt.title(r'approx. of cos(x) at $x_0 = 0$')
plt.grid()
plt.show()
```

Bisher haben wir die Funktion an x0 = 0 approximiert, also eine MacLaurin Reihe gebildet. Das Taylor Polynom hingegen approximiert die funktion f(x) an einer beliebigen Stelle. 
Beispielsweise, wir bilden nun das Taylorpolynom von cos(x) an x = pi, und erweitern das Polynom zustzlich noch um weitere Terme fr eine bessere Annherung.

```python
x = np.linspace(0,2*np.pi,1000)

c0 = -1; c1 = 0; c2 = 1/2; c3 = 0; c4 = -1/24; c5 = 0; c6 = 1/720
P2 = lambda x: c0 + c1*(x-np.pi) + c2*(x-np.pi)**2
P4 = lambda x: c0 + c1*(x-np.pi) + c2*(x-np.pi)**2 + c3*(x-np.pi)**3 + c4*(x-np.pi)**4
P6 = lambda x: c0 + c1*(x-np.pi) + c2*(x-np.pi)**2 + c3*(x-np.pi)**3 + c4*(x-np.pi)**4 + c5*(x-np.pi)**5 + c6*(x-np.pi)**6

plt.plot(x,np.cos(x),'k', label='exact')
plt.plot(x,P2(x),lw=1, label=r'$P_2(x)$')
plt.plot(x, P4(x),lw=1, label=r'$P_4(x)$')
plt.plot(x, P6(x),lw=1, label=r'$P_6(x)$')
ax = plt.gca()
ax.set_ylim([-1.5,1.5])
plt.legend()
plt.title(r'approx. of cos(x) at $x_0 = \pi$')
plt.grid()
plt.show()
```
